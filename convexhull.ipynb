```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
res = (1080, 1920)
empty_img = np.zeros(res, dtype=np.uint8)
plt.imshow(empty_img, cmap='gray')
```

```python
#create binary mask blob
mask1 = np.zeros_like(empty_img)
mask1[500:600, 500:600] = 1

mask2 = np.zeros_like(empty_img)
mask2[550:650, 550:650] = 1

mask3 = np.zeros_like(empty_img)
mask3[600:700, 600:730] = 1

mask = cv2.bitwise_or(mask3, cv2.bitwise_or(mask1, mask2))
plt.imshow(mask, cmap='gray')
```

```python
contours, hierarchy = cv2.findContours(mask, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

contour = max(contours, key = cv2.contourArea)
epsilon = 0.02*cv2.arcLength(contour, True)
poly_approx = cv2.approxPolyDP(contour, epsilon, closed=True)

contour.shape
poly_approx.shape
```
