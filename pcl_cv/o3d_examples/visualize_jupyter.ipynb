```python
import open3d as o3d
import matplotlib.pyplot as plt
```

### render offline
Render without opening GUI, then store snapshot as array

```python
def get_img(geometries):
    vis = o3d.visualization.Visualizer()
    vis.create_window(visible=False) #works for me with False, on some systems needs to be true
    for geometry in geometries:
        vis.add_geometry(geometry)
        vis.update_geometry(geometry)
    vis.poll_events()
    vis.update_renderer()
    img = vis.capture_screen_float_buffer()
    vis.destroy_window()
    return img

```

```python
pcd_path = r'C:\Users\edvar\programming\computer_vision\pcl_cv\data\CSite1_red1-utm.pcd'
pcd = o3d.io.read_point_cloud(pcd_path)
img = get_img([pcd])
plt.imshow(img)
```
